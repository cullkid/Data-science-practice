##### Visualization of data

```python
import pandas as pd
```

```python
dateRange = pd.date_range('2010-01-01', periods=6, freq='YE')
timeSeries = pd.Series([34,54,23,55,78,222], index=dateRange)

timeSeries
```

```python
import matplotlib.pyplot as plt

timeSeries.plot(kind='line',marker='o', color='b',linestyle='--')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid()
plt.show()
```

```python
data = {
    'Year': ['2010-01-01', '2011-01-01', '2012-01-01', '2013-01-01', '2014-01-01', '2015-01-01',],
    'Revenue': [234,564,345,667,887,986]
}

df = pd.DataFrame(data)
df
```

```python
df.plot(x='Year', y='Revenue', marker='o', color='green', legend=False)  #  remove legend to capture it's duty
plt.title('Revenue for the year')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.grid()
plt.show()
```

```python
data = {
    'Cities': ['New York', 'Califonia', 'Gorgia', 'Philadephia', 'Maryland'],
    'Population': [10344,64271,35432,10098,35463]
}

df = pd.DataFrame(data)

df
```

```python
df.sort_values('Cities', ascending=True).plot(x='Cities', y='Population', kind='bar', color='red', legend=False)
plt.title('Population in the major city of USA')
plt.xlabel('Cities')
plt.ylabel('Population')
plt.show()

#  bar plot
```

```python
dfScatter = pd.DataFrame({'A': [1,2,3,4,5,6,7], 'B': [2,4,6,8,5,3,9]})

dfScatter
```

```python
plt.scatter(dfScatter['A'], dfScatter['B'], color='black')
plt.xlabel('A')
plt.ylabel('B')
plt.show()

#  scatter plot that shows the raising from the beggining and later drop drastically before rasining again
#  it scattered everywhere showing por relashionship between them
```
